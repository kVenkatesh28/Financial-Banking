```python
import numpy as np

conditions = [
    (df['Income_Category'] == 'Less than $40K') & (df['Credit_Limit'] < 5000),
    (df['Income_Category'] == 'Less than $40K') & (df['Credit_Limit'] >= 5000),
    (df['Income_Category'].isin(['$40K - $60K', '$60K - $80K'])),
    (df['Income_Category'].isin(['$80K - $120K', '$120K +']))
]

choices = ['High Risk', 'Medium Risk', 'Low Risk', 'Very Low Risk']

df['Risk_Score'] = np.select(conditions, choices, default='Medium Risk')

df[['Income_Category','Credit_Limit','Risk_Score']].head()
```

```python
df['Risk_Score'].value_counts()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

columns = [
    'Customer_Age',
    'Credit_Limit',
    'Total_Revolving_Bal',
    'Avg_Utilization_Ratio',
    'Total_Trans_Amt',
    'Total_Trans_Ct'
]

df[columns] = scaler.fit_transform(df[columns])

df.head()
```

```python
df[columns].describe()
```

```python
df = pd.get_dummies(
    df,
    columns=[
        'Gender',
        'Education_Level',
        'Marital_Status',
        'Income_Category',
        'Card_Category'
    ],
    drop_first=True
)

df.head()
```

```python
df.to_csv("feature_engineered_data.csv", index=False)
```

```python
df.info()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

wcss = []

X = df[['Customer_Age',
        'Credit_Limit',
        'Total_Revolving_Bal',
        'Avg_Utilization_Ratio',
        'Total_Trans_Amt',
        'Total_Trans_Ct']]

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42, n_init=10)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(8,5))
plt.plot(range(1,11), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)

df['Cluster'] = kmeans.fit_predict(X)

df[['Cluster']].head()
```

```python
df['Cluster'].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

sns.scatterplot(
    data=df,
    x='Total_Trans_Amt',
    y='Credit_Limit',
    hue='Cluster',
    palette='Set1'
)


plt.title('Customer Segments using K-Means')
plt.xlabel('Total Transaction Amount')
plt.ylabel('Credit Limit')
plt.show()
```
